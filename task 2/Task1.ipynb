```python
class Email :
    def __init__(self,email):
        self.email = email
        
    def validate_email(self):
        if self.email.count("@") != 1 :
            return "invalid email"
        email_parts = self.email.split("@")
        if "." not in email_parts[1] :
            return "invalid email"
        
        return "valid email"
    def  extract_username(self):
        if self.validate_email() == "invalid email" :
            return "invalid email ... can not extract username"
        at_indix = self.email.index("@")
        return self.email[:at_indix]
    
    def extract_domain(self):
        if self.validate_email() == "invalid email" :
            return "invalid email ... can not extract domain"
        parts = self.email.split("@")
        last_dot_index = parts[1].rfind(".")
        domain = parts[1][:last_dot_index]
        return domain
    
    def  Check_for_Domain_Ending(self):
        if self.validate_email() == "invalid email" :
            return "invalid email ... can not  Check for Domain Ending"
        parts = self.email.split("@")
        last_dot_index = parts[1].rfind(".")
        domain_ending = parts[1][last_dot_index:]
        
        if domain_ending == ".com":
            return "Commercial Domain"
        elif domain_ending == ".edu":
            return "Educational Domain"
        else:
            return "Other Domain"
            
my_email =  Email("Amit_ml@gmail.edu")
print(my_email.validate_email())
print(my_email.extract_username())
print(my_email.extract_domain())
print(my_email.Check_for_Domain_Ending())
```

```python
EncodedMessage = "###!!@mocleW EPGTQ!!!6789"

#  Extract the core part of the message. 
core = ""
for char in EncodedMessage:
    if (char >= 'a' and char <= 'z') or (char >= 'A' and char <= 'Z') or char == ' ':
        core += char 
# split the message to tow words
parts = core.split(" ")
first_word = parts[0]
second_word = parts[1]
#  Reverse the first word: "mocleW" becomes "Welcome". 
reversed_word = first_word[::-1]
#Replace shifted vowels in the second word "EPGTQ" but No vowels to change. 
edited_word = second_word[1:]
# print the final message
print(f"the final message : {reversed_word +' '+ edited_word}")
```
